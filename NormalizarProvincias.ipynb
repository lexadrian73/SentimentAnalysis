```python
import json
import re
import pandas as pd
import numpy as np


f = open('dis_ec.json', encoding="utf8")
  
data = json.load(f)


dict_provincias = []
for k, v in data.items():
    if not (data[k].get('provincia') is None):
        dict_provincias.append({
            'id' : k,
            'provincia' : data[k]['provincia']
        })
        
dict_cantones = []      
for k, v in data.items():
    if not (data[k].get('provincia') is None):
        for k_2, v_2 in data[k]['cantones'].items():
            dict_cantones.append({ 
                'id' : k,
                'provincia' : data[k]['provincia'],
                'canton' : data[k]['cantones'][k_2]['canton']
            })

df = pd.read_csv('tweets_Correa.csv', low_memory=False)
location_column = df['location']

location_list = df['location'].tolist()

location_list=['no ubicacion' if x is np.nan else x for x in location_list]

location_column_regex = []
copy_location = []

for string in location_list:
    if(string != None):
        change_string = re.sub(r'[ ](?=[ ])|[^A-Za-z ]+', ' ', string)
        change_string = change_string.lower().replace('ecuador', '').strip()
        location_column_regex.append(change_string)
        
copy_location = location_column_regex.copy()

for i in range(len(copy_location)):
    if copy_location[i] == '':
        copy_location[i] = 'no ubicacion'
    

for i in range(len(copy_location)):
    for j in dict_cantones:
        if j['canton'].lower() == copy_location[i] or j['canton'].lower() in copy_location[i]:
            copy_location[i] = j['provincia'].lower()
            #print(j['canton'].lower() + " " + copy_location[i])
        
#B
for i in range(len(copy_location)):
    for j in dict_cantones:
        if len(copy_location[i].split()) > 1:
            for c in copy_location[i].split():
                if j['canton'].lower() == c.strip() or j['canton'].lower() in c.strip():
                    copy_location[i] = j['provincia'].lower()
            

df_l = pd.DataFrame({'newlocation': copy_location})
print (df_l)
```
